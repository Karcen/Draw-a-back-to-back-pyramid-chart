```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# 读取数据
excel_file = r'C:\Users\Jiacheng Zheng\Desktop\data.xlsx'  
df1 = pd.read_excel(excel_file, sheet_name='Before 2020')
df2 = pd.read_excel(excel_file, sheet_name='After 2020')

fig = plt.figure(figsize=(8, 10))
gs = gridspec.GridSpec(1,2, width_ratios=[1, 1])
bar_positions = range(len(df1['Labels']))
bar_height = 0.15
ylabelsize = '10'
subtittle = '10'

plt.rcParams['font.family'] = 'Palatino Linotype'
plt.rcParams['font.style'] = 'normal'


ax1 = plt.subplot(gs[0]) 
ax1.invert_xaxis() # 在左侧的图表上绘制，x 轴反转
for i, column in enumerate(['United States', 'Canada', 'Brazil', 'Mexico', 'Argentina']):
    ax1.barh(
        [pos + (i + 1) * bar_height for pos in bar_positions], 
        df1[column],
        height=bar_height,)  
ax1.set_xlabel(' ')
ax1.set_title('Before 2020', fontsize= subtittle)
ax1.set_yticklabels([])   
ax1.tick_params(axis='y', which='both', left=False)


# 上1右1
ax2 = plt.subplot(gs[1])
bar_positions = range(len(df1['Labels']))

for i, column in enumerate(['United States', 'Canada', 'Brazil', 'Mexico', 'Argentina']):
    ax2.barh(
        [pos + i * bar_height for pos in bar_positions],
        df2[column],
        height=bar_height,)
ax2.set_yticks([pos + bar_height * 2 for pos in bar_positions])
ax2.set_yticklabels(df1['Labels'], fontsize=ylabelsize)
ax2.tick_params(axis='y', which='both', left=False)
ax2.set_xlabel(' ')
ax2.set_title('After 2020', fontsize= subtittle)



legend_labels = ['United States', 'Canada', 'Brazil', 'Mexico', 'Argentina']

legend_handles = [
    plt.Rectangle((0, 0), 1, 1, color='#9467bd',),
    plt.Rectangle((0, 0), 1, 1, color='#ff7f0e'),
    plt.Rectangle((0, 0), 1, 1, color='#2ca02c'),
    plt.Rectangle((0, 0), 1, 1, color='#d62728'),
    plt.Rectangle((0, 0), 1, 1, color='#1f77b4'),
]

plt.legend(legend_handles, legend_labels, loc='upper center', bbox_to_anchor=(-0.1, 1.1), prop={'size': 11}, ncol=len(legend_labels))

plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Figure.png', dpi=600)
plt.show()
```
